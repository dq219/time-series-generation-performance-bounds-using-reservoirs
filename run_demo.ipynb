```python
target_amplitde = 10
target_period = 20
target = lambda x: target_amplitde * np.sin(2 * np.pi * x / target_period)

train_duration = 200
test_duration = 200
```

```python
#####################################################################
### The following cell trains using the forgetful FORCE algorithm ###
#####################################################################
```

```python
mode = ['FORCE', 'echo'][0] # choose training mode
gamma = 0.002

network.reset_states(seed = 0)
res_train = network.simulate_train(train_duration, target, mode = mode, γ = gamma)
res_test = network.simulate_test(train_duration + test_duration, target, t_ini = train_duration)
```

```python
train_time = res_train[0]
train_output = res_train[4]

test_time = res_test[0]
test_output = res_test[4]

plt.figure(figsize = (12, 2))
plt.plot(train_time, train_output)
plt.plot(test_time, test_output)
```
